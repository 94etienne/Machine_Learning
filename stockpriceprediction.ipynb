```python
# Remove unwanted columns for cleaning purpose
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
# SETTING UP OUR TARGET FOR MACHINE LEARNING

# Check whether price goes up or down tomorrow
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
sp500
# Tomorrow colum added for prediction from Close Tomorrow get price of next day from Close
```

```python
# based on tomorrow we can set a target
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int) #astype(int) convert into 1 or 0 not True or False
```

```python
sp500
```

```python
sp500 = sp500.loc['1990-01-01':].copy() # taking data only from selected index + 1 date to currect index

sp500
```

```python
# TRAING AN INITIAL MACHINE LEARNING MODEL
from sklearn.ensemble import RandomForestClassifier #(Can not overfit like linear regression)

# Initialize the Random Forest Classifier (TAKE MAX VALUES TO AVOID OVERFIT, random_state=1, IF YOU CONTINUE RUN IT PRODUCE THE SAME RESULTS)
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

# Split the dataset into training and test sets
train = sp500.iloc[:-100]
test = sp500.iloc[-100:]

# Define the predictor variables
predictors = ["Close", "Volume", "Open", "High", "Low"] #COLUMNS NEEDED FOR PREDICTION

# Fit the model to the training data
model.fit(train[predictors], train["Target"]) #FIT OUR TARGET WHICH ONE WE PREDICT
```

```python
# CHECK HOW ACCURATE OUR MODEL
from sklearn.metrics import precision_score #(Check if price goes UP)

# Make predictions using the test set
preds = model.predict(test[predictors])

# Output the predictions
preds

```

```python
import pandas as pd

# Convert the predictions array to a pandas Series with the same index as the test set
preds = pd.Series(preds, index=test.index)
preds
```

```python
precision_score(test["Target"], preds) # 0.54 is some how Good
```

```python
# Plot our prediction
combined = pd.concat([test["Target"], preds], axis=1)
combined.plot()
```

```python
# BUILDING BACK TESTING SYSTEM

def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
import pandas as pd

def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []
    
    # Convert integer-based parameters to time-based if the index is DatetimeIndex
    start_date = data.index[min(start, len(data) - 1)]
    end_date = data.index[-1]
    
    # Generate date ranges for training and testing
    date_ranges = pd.date_range(start=start_date, end=end_date, freq=f'{step}D')
    
    for start_date in date_ranges:
        # Adjust the end_date for the current loop iteration
        end_date = start_date + pd.Timedelta(days=step)
        
        # Select train and test data based on date range
        train = data.loc[:start_date].copy()
        test = data.loc[start_date:end_date].copy()
        
        # Ensure there is enough data to split
        if len(test) > 0:
            predictions = predict(train, test, predictors, model)
            all_predictions.append(predictions)
    
    # Concatenate all predictions
    return pd.concat(all_predictions)

# Example usage
predictions = backtest(sp500, model, predictors)
```

```python

predictions['Predictions'].value_counts()
# PRINT DAY MARKET GOES UP OR DOWN
```

```python

# Calculate the precision score comparing the actual targets and the predictions
precision_score(predictions["Target"], predictions["Predictions"]) 
# OUR ACCURACY IS 52%
```

```python
# ADDING ADDITIONAL PREDICTION TO OUR MODEL
horizons = [2, 5, 60, 250, 1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors += [ratio_column, trend_column]


```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >=.6] = 1 #price >= 60% is UP
    preds[preds <.6] = 0 # price < 60% is UP
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```
